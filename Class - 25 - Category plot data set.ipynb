# IRIS FLOWER DATASET

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#github --- > dataset1
dsiris=pd.read_csv('iris.csv')
print(dsiris)
print(dsiris.head())
sns.set(style="whitegrid")
ax=sns.stripplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='petal length',y='class',data=dsiris)
plt.title('Graph')
plt.show()
```

# Distribution plot

displot : Figure-level interface to distribution plot functions.
histplot : Plot a histogram of binned counts with optional normalization or smoothing.
jointplot : Draw a bivariate plot with univariate marginal distributions.
violinplot : Draw an enhanced boxplot using kernel density estimation.
rugplot:Plot marginal distributions by drawing ticks along the x and y axes.

```python
import seaborn as sns
#Load the data
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
#Create violinplot
sns.violinplot(x="total_bill", data=tips)
#Show the plot
plt.show()
```

```python
#sns.histplot?
sns.histplot(data=tips,x='total_bill',bins=20)
```

```python
#sns.distplot?
sns.distplot(tips["tip"],bins=10)
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],bins=100)
```

```python
sns.distplot(tips["total_bill"],bins=50,kde=False) #(kde  kernel density estimation  pick of choti)
```
